### axes, ticks, and grids

```python
x = range(1,10)
y = [1,2.7,3,4,0,4.3,3,2,1]
```

Axes

```python
fig = plt.figure()    # create a blink figure
ax = fig.add_axes([.1,.1,1,1])    # create axes

ax.plot(x,y)
```

```python
plt.plot(x,y)
```

Ticks

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

Grids

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

### Multiple Charts in One Graph

```python
fig = plt.figure()
fig, mltplt = plt.subplots(1,2)    # 1 row 2 colummns

mltplt[0].plot(x)    # 1st chart
mltplt[1].plot(x,y)    # 2nd chart
```

## labels and annotations

### Labels

The functional method

```python
x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]
plt.bar(x,y)

plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z = [1,2,3,4,.5]

plt.pie(z, labels=['bicycle', 'motorbike', 'car', 'van', 'stroller'])
plt.show()
```

The object-orientated method

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1.6,1])

cars.mpg.plot()

ax.set_xticks(range(32))

ax.grid()

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')
```

### Legend

the functional method

```python
plt.pie(z)
plt.legend(['bicycle', 'motorbike', 'car', 'van', 'stroller'], loc='best')
plt.show()
```

the object-oriented method

```python
# ax.legend(loc='best')    
```

### Annotate

```python
cars.mpg.max()
```

object-orientated method

```python
fig = plt.figure()    # blank figure
ax = fig.add_axes([.1,.1,1.6,1])    # add axes to the blank figure

cars.mpg.plot()    # draw the chart

ax.set_xticks(range(32))    # set tickes on the chart

ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')

ax.set_ylim([0,45])

ax.annotate('Toyota Corolla', xy= (19,33.9), xytext=(21,35),
           arrowprops=dict(facecolor='black', shrink=0.05))
```
